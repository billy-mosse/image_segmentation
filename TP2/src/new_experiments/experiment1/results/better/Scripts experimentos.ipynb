```python
import os
```

```python
d = {}

for filename in os.listdir("."):
    if filename.endswith(".png"):
        print(filename)
        original_filename= filename[filename.rindex("gray_"):]
        name = original_filename[5:-4].replace("_", " ")
        if name in d:
            d[name].append(filename)
        else:
            d[name] = [filename]
```

```python
for name, filenames in d.items():
        filename = filenames[0]
        original_filename= filename[filename.rindex("gray_"):]

        print("\subsubsection{Experimento cualitativo: ", name, "}")
        print("")
        print("Imagen original:")
        print("\includegraphics[scale=1]{Ej2/new_img/%s}" % original_filename)
        print("")
        
        print("Se procesó la imagen de las siguientes maneras:")
        print("\\begin{itemize}")
        for filename in filenames:
            metodo_str = ""
            metodo = int(filename[0])
            if metodo:
                #TODO variar la cantidad de vecinos!!
                metodo_str = "4 nearest neighbors"
            else:
                metodo_str = "Euclidean distance"

            print("\item Método usado para armar el grafo:", metodo_str, "\\\\")
                
            k = float(filename[2:filename.rindex("gray_")-1])
            print("Valor de k:", k)
            print("Resultado:")
            print("\includegraphics[scale=1]{Ej2/new_img/%s}" % filename)
        print("\\end{itemize}")
        
        print("")
```
