```python
def convert_grayscale_image_to_matrix(filename, dst = ""):
    img = Image.open(filename)
    #img.show()
    img_arr = np.array(img)

    if len(img_arr.shape) == 3:
        img_arr = img_arr[:,:,0]

    s = len(filename) - filename.index('.')
    
    w = img_arr.shape[0]
    h = img_arr.shape[1]
    print(dst + filename[:-s] + '.txt')
    np.savetxt(dst + filename[:-s] + '.txt', img_arr, delimiter = ' ', header = "%d %d" % (w, h), comments = '')


def convert_grayscale_image_to_matrix2(filename, dst = ""):
    img = Image.open(filename)
    #img.show()
    img_arr = np.array(img)

    if len(img_arr.shape) == 3:
        img_arr = img_arr[:,:,0]

    s = len(filename) - filename.rindex('.')
    
    w = img_arr.shape[0]
    h = img_arr.shape[1]
    print(filename)
    print(dst + filename[:-s] + '.txt')
    np.savetxt(dst + filename[:-s] + '.txt', img_arr, delimiter = ' ', header = "%d %d" % (w, h), comments = '')
```
