```python
def convert_grayscale_image_to_matrix(filename):
    img = Image.open(filename)
    #img.show()
    img_arr = np.array(img)

    if len(img_arr.shape) == 3:
        img_arr = img_arr[:,:,0]

    s = len(filename) - filename.index('.')
    
    w = img_arr.shape[0]
    h = img_arr.shape[1]
    np.savetxt(filename[:-s] + '.txt', img_arr, delimiter = ' ', header = "%d %d" % (w, h), comments = '')
```

```python
def convert_image_to_grayscale_if_not(filename):
    img = Image.open(filename).convert('LA')
    img.save('gray_' + filename)
```

```python
def mode(colors):
    d = {}
    for color in colors:
        if color not in d:
            d[color]= 1
        else:
            d[color] += 1
    
    max = -1
    color_max = None
    for key, value in d.items():
            if value > max:
                max = value
                color_max = key
    return key
            
def get_most_used_color(cluster, clusters, original_image):
    print("Obteniendo color mas usado del cluster %s" % cluster)
    colors = []
    for i, row in enumerate(clusters):
        for j, cluster2 in enumerate(row):
            if len(cluster) == 0 or len(cluster2) == 0:
                continue
            if int(cluster) == int(cluster2):
                colors.append(int(original_image[i][j]))
                return mode(colors)
d = {}
    
#Mostrar clusters y este retocado en el informe
def convert_cluster_to_most_used_color_in_cluster(clusters, original_image):
    print("Convirtiendo...")
    new_clusters = []
    for row in clusters:
        new_row = []
        for cluster in row:
            if len(cluster) == 0:
                continue
            if cluster not in d.keys():
                print("no")
                d[cluster] = get_most_used_color(cluster, clusters, original_image)
            new_row.append(d[cluster])
        new_clusters.append(new_row)
    
    #O algo asi
    np.savetxt('test.png', new_clusters, delimiter = ' ')
```
